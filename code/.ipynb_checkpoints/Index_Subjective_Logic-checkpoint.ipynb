```python
import math

def comb(n,m):
    if(n < m):
        return 0
    else:
        return math.factorial(n)/(math.factorial(m)*math.factorial(n-m))

def index(composite, cardinality):
    """Receives a list (composite) with elements in range(0,cardinality)
    and returns its index according to the rules of Subjective Logic (but zero based).
    If a factorial vector is allowed the complexity will go from
    O(n^2) to O(n). But it doesn't make a lot of diference since cardinality 
    can't be greater than 70."""
    
    ceil = 0
    for i in range(1,len(composite)+1):
        ceil += comb(cardinality, i)
    
    left = len(composite)
    for element in composite:
        pos_elem = cardinality - element - 1
        ceil -= comb(pos_elem, left)
        left -= 1
    
    return ceil-1

def inv_index(index,cardinality):
    """Receives a 0-based index of composite values in subjective logic and
    the cardinality of the domain and returns it's 0-based composite value."""
    index += 1
    composite = []
    n_elem = 1
    while(True):
        aux = comb(cardinality,n_elem)
        if(aux < index):
            index -= aux
            n_elem += 1
        else:
            break

    pos_choosing = 0
    while(len(composite) < n_elem):
        aux = comb(cardinality - pos_choosing - 1, n_elem - len(composite)-1)
        
        if(n_elem - len(composite) == 1):
            composite.append(int(pos_choosing + index - 1))
        elif(aux < index):
            index -= aux
            pos_choosing += 1
        else:
            composite.append(pos_choosing)
            pos_choosing += 1
    return composite
```
